### Chart Title and Legend

```python
plt.plot(years, apples);
plt.plot(years, oranges);

plt.xlabel('Years');
plt.ylabel('Yield (tons per hectare)');

plt.title('Yield of Apples vs Oranges In Konto')
plt.legend(['Apples', 'Oranges']);
```

### Line Markers In Matplotlib

```python
plt.plot(years, apples, marker='o');
plt.plot(years, oranges, marker='x');

plt.xlabel('Years');
plt.ylabel('Yield (tons per hectare)');

plt.title('Yield of Apples vs Oranges In Konto')
plt.legend(['Apples', 'Oranges']);
```

### Styline lines and markers

1. color or c: set color of the line
2. linestyle or ls: set line styles between dashed or solid
3. linewidth or lw: set the width of a line
4. markersize or ms: set the size or markers
4. markeredgecolor or mec: set the edge color for markers
6. markeredgewidth or mew: set the edge width for markers
7. markerfacecolor or mfc: set the fill color for markers
8. alpha: opacity of plot

```python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, mew=3, alpha=.5)
plt.xlabel('Years');
plt.ylabel('Yield (tons per hectare)');

plt.title('Yield of Apples vs Oranges In Konto')
plt.legend(['Apples', 'Oranges']);
```

### using fmt for linestyle, marker and linecolor
    fmt = [line_style][marker][line_color]

```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r', alpha=.5)
plt.xlabel('Years');
plt.ylabel('Yield (tons per hectare)');

plt.title('Yield of Apples vs Oranges In Konto')
plt.legend(['Apples', 'Oranges']);
```

## Scatter Plot

1. values of 2 variables are plotted as points on 2-D grid
2. additionally, can use 3rd variable to determine the size and color of the points

```python
# load irish flower dataset in pandas dataframe
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df.head()
```

```python
flowers_df.species.unique()
```

```python
# checking relationship between sepal_width and sepal_length
plt.plot(flowers_df['sepal_length'], flowers_df['sepal_width']);
```

```python
# oops a bad idea to plot with line plot, use scatter plot from seaborn
sns.scatterplot(flowers_df['sepal_length'], flowers_df['sepal_width']);
```

## Adding hues to distinct clusters with some outliers

```python
sns.scatterplot(
    flowers_df['sepal_length'],
    flowers_df['sepal_width'],
    hue=flowers_df.species,
    s=100,
);
```

##  Customizing Seaborn Figures

```python
# as seaborn is derived from matplotlib use can use same customizing as matplotlib
plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')

sns.scatterplot(
    flowers_df['sepal_length'],
    flowers_df['sepal_width'],
    hue=flowers_df.species,
    s=100,
);
```

## Plotting using pandas dataframes

```python
plt.title('Sepal Dimensions')
sns.scatterplot('sepal_length', 'sepal_width', hue='species', s=100, data=flowers_df);
```

## Histogram

1. represent the visualization of data by forming bins along the range of data
2. draws bars to show no of observations

```python
flowers_df = sns.load_dataset("iris")
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width);
```

### Controling the size, interval (ie bins in statistical terms)

```python
plt.hist(flowers_df.sepal_width, bins=10);
```

```python
import numpy as np
np.arange(2, 5, 0.25)
```

```python
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25));
```

```python
# bins of unequal sizes
plt.hist(flowers_df.sepal_width, bins=[1, 3, 4, 4, 5]);
```

## Multiple Histogram in single chart

```python
setosa_df = flowers_df[flowers_df.species=='setosa']
versicolor_df = flowers_df[flowers_df.species=='versicolor']
virginica_df = flowers_df[flowers_df.species=='virginica']
```

```python
plt.hist(setosa_df.sepal_width, bins=np.arange(2, 5, 0.25), alpha=0.45);
plt.hist(versicolor_df.sepal_width, bins=np.arange(2, 5, 0.25), alpha=0.4);
plt.legend(['Setosa', 'VersiColor']);
```

```python
# convert overlapping hist to stacked hist
plt.title('Distribution of Sepal width');
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins=np.arange(2, 5, 0.25), stacked=True);
plt.legend(['Setosa', 'VersiColor', 'verginicia']);
```

# Multiple Charts in grid 

### plt.subplots

```python
# set number of plots to include
plt.subplots(2,3);

# set spacing between graphs
plt.tight_layout(pad=2)
```

```python
fig, axes = plt.subplots(2,3, figsize=(16, 8))
axes[0,0].plot(years, apples, 'o--r')
axes[0,0].plot(years, oranges, 's-b')
axes[0,0].set_xlabel('Apples')
axes[0,0].set_xlabel('Oranges')
axes[0,0].set_title('Yield of Oranges');
axes[0,0].legend(['Apples', 'Oranges'])
plt.tight_layout(pad=2)

# using sns needs passing ax as axes
axes[0,1].set_title('Sepal length vs. width')
sns.scatterplot(flowers_df.sepal_length,
               flowers_df.sepal_width,
               hue=flowers_df.species,
               s=100,
               ax=axes[0,1])

# Use the axes for plotting
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
              bins=np.arange(2, 5, 0.25),
              stacked=True);
axes[0,2].legend(['setosa', 'versicolor', 'virginica'])

# pass the axis into seaborn
axes[1,0].set_title('Restaurant bills')
sns.barplot('day', 'total_bill', hue='sex', data=tips_df, ax=axes[1,0]);

# pass the axis into seaborn
axes[1,1].set_title('Fligth Traffic')
sns.heatmap(flights_df, cmap='Blues', ax=axes[1,1]);

# plot an image
axes[1,2].set_title('Data Science Meme')
axes[1,2].imshow(img);
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])

plt.tight_layout(pad=2);
```
